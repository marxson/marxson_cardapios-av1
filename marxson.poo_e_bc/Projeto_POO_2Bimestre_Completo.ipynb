## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Classe principal Prato
class Prato:
    # Construtor (__init__) para inicializar os atributos da classe
    def __init__(self, nome, ingredientes, calorias):
        self.nome = nome              # Atributo 1: nome do prato
        self.ingredientes = ingredientes  # Atributo 2: ingredientes do prato
        self.calorias = calorias        # Atributo 3: calorias do prato

    # M√©todo para exibir os dados do prato
    def exibir_dados(self):
        print(f"Prato: {self.nome}")
        print(f"Ingredientes: {self.ingredientes}")
        print(f"Calorias: {self.calorias} kcal")

    # M√©todo de a√ß√£o 1: Atualizar calorias
    def atualizar_calorias(self, novas_calorias):
        self.calorias = novas_calorias
        print(f"As calorias do prato {self.nome} foram atualizadas para {self.calorias} kcal.")

    # M√©todo de a√ß√£o 2: Adicionar ingredientes
    def adicionar_ingredientes(self, novos_ingredientes):
        self.ingredientes += ", " + novos_ingredientes
        print(f"Novos ingredientes adicionados ao prato {self.nome}: {novos_ingredientes}")

# Subclasse PratoVegetariano herda de Prato
class PratoVegetariano(Prato):
    # Construtor da subclasse (usando super() para herdar da classe pai)
    def __init__(self, nome, ingredientes, calorias, tipo_vegetariano):
        super().__init__(nome, ingredientes, calorias)  # Herdando atributos da classe Pai
        self.tipo_vegetariano = tipo_vegetariano  # Novo atributo exclusivo da subclasse

    # Novo m√©todo exclusivo da subclasse
    def exibir_tipo_vegetariano(self):
        print(f"O prato {self.nome} √© do tipo vegetariano: {self.tipo_vegetariano}")

# Criando objetos da classe Prato
prato1 = Prato("Arroz com Feij√£o", "Arroz, Feij√£o, Carne, Salada", 450)
prato2 = Prato("Macarr√£o √† Bolonhesa", "Macarr√£o, Carne Mo√≠da, Molho de Tomate, Queijo", 600)

# Criando objetos da subclasse PratoVegetariano
prato_vegetariano1 = PratoVegetariano("Feijoada Vegetariana", "Feij√£o, Tofu, Legumes", 350, "Vegana")
prato_vegetariano2 = PratoVegetariano("Lasanha de Abobrinha", "Abobrinha, Queijo, Molho de Tomate, Espinafre", 400, "Vegetariana")

# Exibindo dados dos pratos
print("\nDetalhes do prato 1 (Prato comum):")
prato1.exibir_dados()

print("\nDetalhes do prato 2 (Prato comum):")
prato2.exibir_dados()

# Atualizando dados do prato
prato1.atualizar_calorias(500)
prato2.adicionar_ingredientes("Azeitona")

# Exibindo dados dos pratos vegetarianos
print("\nDetalhes do prato 1 (Prato Vegetariano):")
prato_vegetariano1.exibir_dados()
prato_vegetariano1.exibir_tipo_vegetariano()

print("\nDetalhes do prato 2 (Prato Vegetariano):")
prato_vegetariano2.exibir_dados()
prato_vegetariano2.exibir_tipo_vegetariano()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
class Prato:
    # Construtor (__init__) para inicializar os atributos da classe
    def __init__(self, nome, ingredientes, calorias):
        self.nome = nome               # Atributo 1: nome do prato
        self.ingredientes = ingredientes # Atributo 2: ingredientes do prato
        self.__calorias = calorias      # Atributo 3: calorias (privado)

    # M√©todos get e set para o atributo privado __calorias
    def get_calorias(self):
        return self.__calorias

    def set_calorias(self, novas_calorias):
        if novas_calorias > 0:  # Validando se as calorias s√£o positivas
            self.__calorias = novas_calorias
            print(f"As calorias do prato {self.nome} foram atualizadas para {self.__calorias} kcal.")
        else:
            print("Valor de calorias inv√°lido! Deve ser maior que zero.")

    # M√©todo de l√≥gica interna: Verificar se o prato √© saud√°vel
    def verificar_saudavel(self):
        if self.__calorias <= 500:
            print(f"O prato {self.nome} √© saud√°vel.")
        else:
            print(f"O prato {self.nome} n√£o √© saud√°vel.")

    # M√©todo para exibir os dados do prato
    def exibir_dados(self):
        print(f"Prato: {self.nome}")
        print(f"Ingredientes: {self.ingredientes}")
        print(f"Calorias: {self.get_calorias()} kcal")

# Criando objetos da classe Prato
prato1 = Prato("Arroz com Feij√£o", "Arroz, Feij√£o, Carne, Salada", 450)
prato2 = Prato("Macarr√£o √† Bolonhesa", "Macarr√£o, Carne Mo√≠da, Molho de Tomate, Queijo", 600)

# Exibindo dados do prato
print("\nDetalhes do prato 1 (Prato comum):")
prato1.exibir_dados()

print("\nDetalhes do prato 2 (Prato comum):")
prato2.exibir_dados()

# Alterando calorias utilizando o set
prato1.set_calorias(550)  # Atualizando com um valor v√°lido
prato1.set_calorias(-100)  # Tentando atualizar com um valor inv√°lido

# Verificando se o prato √© saud√°vel
prato1.verificar_saudavel()
prato2.verificar_saudavel()
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3

# --------------------------------------------------------------
# Fun√ß√£o para conectar ao banco de dados SQLite
def conectar_banco():
    return sqlite3.connect('cardapio_escolar.db')

# --------------------------------------------------------------
# Fun√ß√£o para criar a tabela no banco de dados
def criar_tabela():
    conn = conectar_banco()
    cursor = conn.cursor()
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS pratos (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        nome TEXT NOT NULL,
        ingredientes TEXT NOT NULL,
        calorias INTEGER NOT NULL,
        saudavel BOOLEAN NOT NULL
    )
    ''')
    conn.commit()
    conn.close()

# --------------------------------------------------------------
# Classe principal Prato
class Prato:
    def __init__(self, nome, ingredientes, calorias):
        # Atributos de inst√¢ncia
        self.nome = nome
        self.ingredientes = ingredientes
        self.__calorias = calorias  # Atributo privado

    # M√©todos get e set para o atributo privado __calorias
    def get_calorias(self):
        return self.__calorias

    def set_calorias(self, novas_calorias):
        if novas_calorias > 0:  # Validando as calorias
            self.__calorias = novas_calorias
            print(f"As calorias do prato '{self.nome}' foram atualizadas para {self.__calorias} kcal.")
        else:
            print("Valor de calorias inv√°lido! Deve ser maior que zero.")

    # M√©todo para verificar se o prato √© saud√°vel
    def verificar_saudavel(self):
        return self.__calorias <= 500

    # M√©todo para exibir dados do prato (polimorfismo ser√° aplicado na subclasse)
    def exibir_dados(self):
        print(f"\nPrato: {self.nome}")
        print(f"Ingredientes: {self.ingredientes}")
        print(f"Calorias: {self.__calorias} kcal")
    
    # M√©todo para inserir prato no banco de dados
    def inserir_no_banco(self):
        conn = conectar_banco()
        cursor = conn.cursor()
        cursor.execute('''
        INSERT INTO pratos (nome, ingredientes, calorias, saudavel)
        VALUES (?, ?, ?, ?)
        ''', (self.nome, self.ingredientes, self.__calorias, self.verificar_saudavel()))
        conn.commit()
        conn.close()
        print(f"\nPrato '{self.nome}' inserido no banco de dados.")

    # M√©todo para excluir prato do banco de dados
    @staticmethod
    def excluir_do_banco(prato_id):
        conn = conectar_banco()
        cursor = conn.cursor()
        cursor.execute('DELETE FROM pratos WHERE id = ?', (prato_id,))
        conn.commit()
        conn.close()
        print(f"\nPrato com ID {prato_id} exclu√≠do do banco de dados.")

    # M√©todo para consultar pratos no banco de dados
    @staticmethod
    def consultar_no_banco():
        conn = conectar_banco()
        cursor = conn.cursor()
        cursor.execute('SELECT * FROM pratos')
        pratos = cursor.fetchall()
        if pratos:
            for prato in pratos:
                print(f"\nID: {prato[0]}, Nome: {prato[1]}, Ingredientes: {prato[2]}, Calorias: {prato[3]}, Saud√°vel: {'Sim' if prato[4] else 'N√£o'}")
        else:
            print("\nNenhum prato encontrado no banco de dados.")
        conn.close()

# --------------------------------------------------------------
# Fun√ß√£o para cadastrar um novo prato
def cadastrar_prato():
    print("\n--- Cadastro de Novo Prato ---")
    nome = input("Digite o nome do prato: ")
    ingredientes = input("Digite os ingredientes do prato: ")
    calorias = int(input("Digite a quantidade de calorias do prato: "))

    prato = Prato(nome, ingredientes, calorias)
    prato.inserir_no_banco()

# --------------------------------------------------------------
# Fun√ß√£o para consultar pratos por calorias
def consultar_pratos_por_calorias():
    print("\n--- Consulta de Pratos por Calorias ---")
    faixa_calorias_min = int(input("Digite o valor m√≠nimo de calorias: "))
    faixa_calorias_max = int(input("Digite o valor m√°ximo de calorias: "))
    
    conn = conectar_banco()
    cursor = conn.cursor()
    cursor.execute('''
    SELECT * FROM pratos WHERE calorias BETWEEN ? AND ?
    ''', (faixa_calorias_min, faixa_calorias_max))
    
    pratos = cursor.fetchall()
    if pratos:
        for prato in pratos:
            print(f"\nID: {prato[0]}, Nome: {prato[1]}, Ingredientes: {prato[2]}, Calorias: {prato[3]}, Saud√°vel: {'Sim' if prato[4] else 'N√£o'}")
    else:
        print("\nNenhum prato encontrado dentro dessa faixa de calorias.")
    conn.close()

# --------------------------------------------------------------
# Fun√ß√£o para excluir um prato
def excluir_prato():
    print("\n--- Exclus√£o de Prato ---")
    prato_id = int(input("Digite o ID do prato que deseja excluir: "))
    Prato.excluir_do_banco(prato_id)

# --------------------------------------------------------------
# Fun√ß√£o para atualizar as calorias de um prato
def atualizar_calorias():
    print("\n--- Atualiza√ß√£o de Calorias de Prato ---")
    prato_id = int(input("Digite o ID do prato que deseja atualizar: "))
    novas_calorias = int(input("Digite a nova quantidade de calorias: "))
    
    conn = conectar_banco()
    cursor = conn.cursor()
    cursor.execute('''
    UPDATE pratos SET calorias = ? WHERE id = ?
    ''', (novas_calorias, prato_id))
    conn.commit()
    conn.close()
    print(f"\nCalorias do prato de ID {prato_id} atualizadas para {novas_calorias} kcal.")

# --------------------------------------------------------------
# Fun√ß√£o para testar o sistema com entradas do usu√°rio
def testar_sistema():
    while True:
        print("\nEscolha uma op√ß√£o:")
        print("1. Cadastrar novo prato")
        print("2. Consultar pratos por calorias")
        print("3. Excluir prato")
        print("4. Atualizar calorias de um prato")
        print("5. Consultar todos os pratos")
        print("6. Sair")

        opcao = int(input("Digite o n√∫mero da op√ß√£o desejada: "))

        if opcao == 1:
            cadastrar_prato()
        elif opcao == 2:
            consultar_pratos_por_calorias()
        elif opcao == 3:
            excluir_prato()
        elif opcao == 4:
            atualizar_calorias()
        elif opcao == 5:
            Prato.consultar_no_banco()
        elif opcao == 6:
            print("Saindo...")
            break
        else:
            print("\nOp√ß√£o inv√°lida! Tente novamente.")

# --------------------------------------------------------------
# Fun√ß√£o principal para executar o sistema
if __name__ == "__main__":
    # Criando a tabela no banco de dados
    criar_tabela()

    # Testando a funcionalidade do sistema com inputs interativos
    testar_sistema()
```
